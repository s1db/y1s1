## Building a better flashlight

<img src="imgs/flashlight.jpg" width="60%">


## Design brief
You are tasked to design a new input technique for a flashlight with several operating modes.  The flashlight is a new design which includes a single "touch slider" interface on the barrel of the flashlight. It does not have any buttons at all. The multiple modes controlled with a touch control that goes beyond a simple "on/off" switch presents an interaction challenge: how to control the different states of the flashlight given potentially noisy, unreliable input.  

<img src="imgs/flashlight_design.png" width="60%">

*The flashlight, with a single one-dimensional slider control for all functions.*

Users should be able to easily and accurately access all flashlight functions easily and quickly in challenging situations. It might be used in emergency situations, or in the dark, or where users struggle to grasp the flashlight (e.g. when wearing gloves on a very cold day), or in situations where control is very limited or there is lots of external disturbance (imagine trying to operate the light in a very turbulent plane descending uncontrollably in darkness). As more functionality is added, making the flashlight usable becomes challenging.

### Flashlight input

The only control is via the touch slider, which can detect forward and backwards sliding motions of the user's thumb across a touch sensitive strip on the side of the torch.  Sliding motions can go up an down at different speeds.  Input actions can be implemented as a single slide movement or a combination of slides.  

For your implementation, the slider will be simulated using the mouse.

## Designing states and functions

### Flashlight functionality

Your flashlight must support at least the following functions:
* Off
* On, Low Light Level
* On, High Light Level
* On, Flashing Light Level (SOS mode)

Optional additional functionality:
* Locked in Off State



## Task: Develop and sketch interaction designs
Individually, brainstorm and discuss how a touch slider based flashlight might work. You might want to consider different *scenarios*, like those described above or different *personas* who might use the device:

* Anne, an aircraft pilot and keeps a torch in case of instrument malfunction;
* Andrew, who keeps a torch in a drawer in the hall for the power going out;
* Anjie, age 6, who is frightened of the dark; 
* Abel who is a night fisherman in Alaska and needs to hook lines in freezing darkness. 

Invent your own scenarios or personas.

Develop some conceptual designs for operating the flashlight. Come up with several different possibilities for how the flashlight might work; bear in mind that the light may be used where fast access is required; where users may be stressed or under time pressure; where users are not familiar with the design of the device; where the user's hands are cold, covered with gloves or subject to vibrations that might limit control abilities.

Before you begin implementing the flashlight interface, choose one of your designs and formalise it using a **state machine diagram** that defines how your flashlight will work in terms of how it switches from one state to another as the input changes.  
During this lab session, sketch state diagrams that represent at least two different designs that capture all of the required functionality described above.  This can be completed on paper or using a digital sketching tool (e.g. [draw.io](http://draw.io)), but must be saved for reference for the following labs.

### States
Represent each **state** as a circle in your state machine diagram.  All of the required output states like "on low" or "off" above must be included in your design. You might choose to include intermediate states in your interface.

### Transitions
Represent each **transition** as an arrow connecting two states, labeled clearly with the action that will trigger the transition. This should be some kind of up/down motion on the slider sensor.

The key aspect of your design is how you implement users' actions to transition between states.  There are numerous ways to divide the sensing area, incorporate thresholds, and combine different actions to trigger the state transitions.  Consider how you map input to transitions carefully and ensure your choices are motivated by usability heuristics.  

Keep in mind that your assessment will include implementing these transitions using mouse input to simulate sliding actions.  Your final design can change from your initial design, but the implementation will be easier if you consider usability from the beginning.

### Timeouts
You may also need to consider how you will implement *timeouts* in the design, as well as explicit user controls made by sliding the fingers.  For example, the flashlight might reset to the "off" mode if it is not touched for 60 seconds (this probably isn't a good design choice!). Include any such timeouts in your state machine diagram.

Based on these discussions, you may chose to improve and iterate on your personas and scenarioes in preparation for next week's lab.
